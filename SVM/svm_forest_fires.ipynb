```python
fire = pd.read_csv("forestfires.csv")
fire.head()
```

```python
string_col = ['month','day','size_category']
```

```python
# To convert string column to numeric
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in string_col:
    fire[i] = label_encoder.fit_transform(fire[i])
```

```python
fire.head()
```

```python
X = fire.iloc[:,:10]
X.head()
```

```python
y = fire["size_category"]
y.head()
```

```python
# SVM Classification using kernels: linear, poly, rbf
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# Split the Data intp Training Data and Test Data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
#  Using Kernel as linear type

model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
train_pred_lin = model_linear.predict(X_train)
test_pred_lin = model_linear.predict(X_test)

metrics.accuracy_score(y_test, model_linear.predict(X_test))
```

```python
confusion_matrix(y_test, model_linear.predict(X_test))
```

```python
#  Using Kernel as poly type

model_poly = SVC(kernel='poly')
model_poly.fit(X_train, y_train)
train_pred_poly = model_poly.predict(X_train)
test_pred_poly = model_poly.predict(X_test)

metrics.accuracy_score(y_test, model_poly.predict(X_test))
```

```python
#  Using Kernel as rbf type

model_rbf = SVC(kernel='rbf')
model_rbf.fit(X_train, y_train)
train_pred_rbf = model_rbf.predict(X_train)
test_pred_rbf = model_rbf.predict(X_test)

metrics.accuracy_score(y_test, model_rbf.predict(X_test))
```

```python
#  Using Kernel as sigmoid type

model_sig = SVC(kernel='sigmoid')
model_sig.fit(X_train, y_train)
train_pred_sig = model_sig.predict(X_train)
test_pred_sig = model_sig.predict(X_test)

metrics.accuracy_score(y_test, model_sig.predict(X_test))
```
